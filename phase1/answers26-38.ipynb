26. **Checking if a Number is a Narcissistic Number**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, Number Theory  
    **Description**: Write a program to check if a number is a narcissistic number (where the sum of its digits raised to the power of the number of digits equals the number itself).  
    **Example**:  
    Input: `number = 153`  
    Output: `Narcissistic Number`  
    Explanation: 153 is a narcissistic number because 1^3 + 5^3 + 3^3 = 153. 

````python
# Time Complexity: O(d) where d is number of digits
# Space Complexity: O(d) for string conversion
def narcissistic(num):
    digits = str(num)
    power = len(digits)
    res = 0
    for i in range(len(digits)):
        res += int(digits[i]) ** power
    return res == num
    
print(narcissistic(153))
````

````python
# Pythonic Solution
# Time: O(d), Space: O(d)
def narcissistic_pythonic(num):
    digits = str(num)
    return sum(int(d) ** len(digits) for d in digits) == num

print(narcissistic_pythonic(153))
````

27 .**Generating a Pattern of Numbers**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, Patterns  
    **Description**: Write a program to generate number patterns (e.g., sequential numbers in a matrix).  
    **Example**:  
    Input: `rows = 3`  
    Output:  
    ```
    1  
    2 3  
    4 5 6  
    ```  
    Explanation: A number pattern with 3 rows is generated.

````python
# Time Complexity: O(n^2) where n is number of rows
# Space Complexity: O(1)
def patternNum(rows):
    a = 0
    for i in range(1,rows+1):
        for j in range(i):
            a+=1
            print(a, end ="")
        print()
        
patternNum(3)
patternNum(5)
````

````python
# Pythonic Solution
# Time: O(n^2), Space: O(1)
def patternNum_pythonic(rows):
    num = 1
    for i in range(1, rows + 1):
        print(''.join(str(num + j) for j in range(i)))
        num += i

patternNum_pythonic(3)
````

28. **Finding the Sum of the Digits of the Factorial of a Number**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, Mathematical Computations  
    **Description**: Write a program to find the sum of the digits of the factorial of a given number.  
    **Example**:  
    Input: `number = 4`  
    Output: `9`  
    Explanation: The factorial of 4 is 24, and the sum of the digits (2 + 4) is 6.  

````python
# Time Complexity: O(n + d) where n is input, d is digits in factorial
# Space Complexity: O(d) for string conversion
def factorialSum(num):
    res = num
    while num > 1:
        res = res*(num-1)
        num -= 1
    
    print(res)
    digit = str(res)
    sum = 0
    for i in range(len(digit)):
        sum += int(digit[i])
    return sum
        
print(factorialSum(4))
````

````python
# Pythonic Solution using math.factorial
# Time: O(n + d), Space: O(d)
from math import factorial

def factorialSum_pythonic(num):
    return sum(int(d) for d in str(factorial(num)))

print(factorialSum_pythonic(4))
````

29. **Finding the Largest Palindrome in a String**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, String Manipulation  
    **Description**: Write a program to find the largest palindrome in a given string.  
    **Example**:  
    Input: `string = "babad"`  
    Output: `"bab"` or `"aba"`  
    Explanation: Both "bab" and "aba" are valid palindromes in the string.  

````python
# Brute Force: Time O(n^3), Space O(n)
def largepalindromsub(word):
    res = ""
    for i in range(len(word)):
        for j in range(i, len(word)):
            sub = word[i:j+1]
            if sub == sub[::-1] and len(sub) > len(res):
                res = sub
    return res
print(f"fun1 {largepalindromsub("babad")}")

def largepalindromsub2(word):
    n = len(word)
    max_length = 1
    start = 0
    for i in range(n):
        for j in range(i,n):
            low, high = i,i+j
            while low > 0 and high < n and word[low] == word[high]:
                currlen = high-low+1
                if currlen > max_length:
                    start = low
                    max_length = currlen
                low -= 1
                high +=1
    return word[start:start+max_length]

print(largepalindromsub2("babad"))

def check(word, low, high):
    while low < high:
        if word[low] != word[high]:
            return False
        low += 1
        high -= 1
    return True
        
def largepalindromsub3(word):
    n = len(word)
    max_length = 1
    start = 0
    for i in range(n):
        for j in range(i,n):
            if check(word, i, j) and j-i+1 > max_length:
                start = i
                max_length = j-i+1
    return word[start:start+max_length]

print(largepalindromsub3("babad"))
    
````

````python
# Optimized: Expand Around Center
# Time: O(n^2), Space: O(1)
def expand_center(s, left, right):
    while left >= 0 and right < len(s) and s[left] == s[right]:
        left -= 1
        right += 1
    return s[left + 1:right]

def largepalindrome_optimized(word):
    result = ""
    for i in range(len(word)):
        odd = expand_center(word, i, i)
        even = expand_center(word, i, i + 1)
        result = max(result, odd, even, key=len)
    return result

print(largepalindrome_optimized("babad"))
````

````python
# Pythonic Solution
# Time: O(n^2), Space: O(1)
def largepalindrome_pythonic(s):
    def expand(l, r):
        while l >= 0 and r < len(s) and s[l] == s[r]:
            l, r = l - 1, r + 1
        return s[l + 1:r]
    
    return max((expand(i, i) for i in range(len(s))), key=len) if s else ""

print(largepalindrome_pythonic("babad"))
````

30. **Finding Missing Numbers in a Sequence**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, Arrays  
    **Description**: Write a program to find missing numbers in a sequence from 1 to n.  
    **Example**:  
    Input: `sequence = [1, 2, 4, 5]`  
    Output: `[3]`  
    Explanation: The missing number in the sequence from 1 to 5 is 3.  

````python
# Time Complexity: O(n) where n is length of array
# Space Complexity: O(1)
def missingNum(nums):
    n = len(nums)
    total_sum = n * (n + 1) // 2
    return total_sum - sum(nums)
print(missingNum([0,1,3]))

def missing2(nums):
    start = nums[0]
    for i in range(1,len(nums)):
        if nums[i] != i+start:
            return i+start
    return -1
print(missing2([0,1,3]))
````

````python
# Optimized: XOR approach
# Time: O(n), Space: O(1)
def missingNum_xor(nums):
    xor = len(nums)
    for i, num in enumerate(nums):
        xor ^= i ^ num
    return xor

print(missingNum_xor([0,1,3]))

# Pythonic: Set difference
# Time: O(n), Space: O(n)
def missingNum_pythonic(nums):
    return (set(range(len(nums) + 1)) - set(nums)).pop()

print(missingNum_pythonic([0,1,3]))
````

    ```

    ```  
    Explanation: Pascal's Triangle with 4 rows is generated.

````python
# Time Complexity: O(n^2) where n is number of rows
# Space Complexity: O(n^2) for storing triangle
def pascal(rows):
    result = [[1]]
    
    for i in range(rows-1):
        temp = []
        temp.append(1)
        temp.insert(-1, 1)
        last = result[-1]
        l = 1
        for j in range(len(last)-1):
            temp.insert(l, last[j] + last[j+1])
            l +=1
        result.append(temp)
    return result

# print(pascal(5))
def print_pascal(rows):
    result = pascal(rows)
    for row in result:
        print(' '.join(str(num) for num in row))
        # print(row)
print_pascal(5)
            
````

````python
# Optimized Solution
# Time: O(n^2), Space: O(n^2)
def pascal_optimized(rows):
    triangle = [[1]]
    for i in range(1, rows):
        row = [1]
        for j in range(1, i):
            row.append(triangle[i-1][j-1] + triangle[i-1][j])
        row.append(1)
        triangle.append(row)
    return triangle

for row in pascal_optimized(5):
    print(' '.join(map(str, row)))

# Pythonic Solution
# Time: O(n^2), Space: O(n^2)
def pascal_pythonic(rows):
    triangle = [[1]]
    for _ in range(rows - 1):
        triangle.append([1] + [triangle[-1][i] + triangle[-1][i+1] for i in range(len(triangle[-1])-1)] + [1])
    return triangle

for row in pascal_pythonic(5):
    print(' '.join(map(str, row)))
````

32. **Finding the Median of an Array**  
    **Difficulty**: Medium  
    **Topics**: Arrays, Sorting  
    **Description**: Write a program to find the median of an array of numbers.  
    **Example**:  
    Input: `array = [3, 1, 2, 4, 5]`  
    Output: `3`  
    Explanation: The median of the sorted array [1, 2, 3, 4, 5] is 3. 

````python
# Time Complexity: O(n log n) due to sorting
# Space Complexity: O(1) if in-place sort, O(n) otherwise
def median(nums):
    nums.sort()
    n = len(nums)
    if n%2 == 1:
        return nums[n//2]
    else:
        return (nums[n//2 - 1] + nums[n//2]) / 2

print(median([3,1,6,2,5,4]))
````

````python
# Optimized: Quickselect (average O(n))
# Time: O(n) average, O(n^2) worst, Space: O(log n)
import random

def quickselect(nums, k):
    pivot = random.choice(nums)
    left = [x for x in nums if x < pivot]
    mid = [x for x in nums if x == pivot]
    right = [x for x in nums if x > pivot]
    if k < len(left):
        return quickselect(left, k)
    elif k < len(left) + len(mid):
        return mid[0]
    return quickselect(right, k - len(left) - len(mid))

def median_optimized(nums):
    n = len(nums)
    if n % 2 == 1:
        return quickselect(nums, n // 2)
    return (quickselect(nums, n // 2 - 1) + quickselect(nums, n // 2)) / 2

print(median_optimized([3,1,6,2,5,4]))

# Pythonic Solution
# Time: O(n log n), Space: O(n)
from statistics import median as stat_median

def median_pythonic(nums):
    return stat_median(nums)

print(median_pythonic([3,1,6,2,5,4]))
````
