12. **Counting Vowels and Consonants in a String**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, String Manipulation  
    **Description**: Write a program to count vowels and consonants in a given string.  
    **Example**:  
    Input: `string = "hello world"`  
    Output: `Vowels: 3, Consonants: 7`  
    Explanation: "hello world" contains 3 vowels (e, o, o) and 7 consonants (h, l, l, w, r, l, d).  


### 📝 Your Solutions vs Optimized Comparison

```python
# ========== YOUR SOLUTIONS ==========

# Your Solution 1: Two pointers (creative but unnecessary)
def countVowelsConsonants(s):
    vowels = 'aeiouAEIOU'
    vowel_count = 0
    consonant_count = 0
    
    start = 0
    end = len(s) - 1
    while start <= end:
        if s[start] in vowels:
            vowel_count += 1
        elif s[start].isalpha():
            consonant_count += 1
        
        if start != end:  # ⚠️ Extra complexity
            if s[end] in vowels:
                vowel_count += 1
            elif s[end].isalpha():
                consonant_count += 1
        start += 1
        end -= 1
    return vowel_count, consonant_count

# Time: O(n) - visits each character once
# Space: O(1) - only counters and pointers
# ⚠️ Two pointers don't help here - simple loop is clearer

# Your Solution 2: Simple loop ⭐ BEST approach!
def countVowels(s):
    vowels = 'aeiouAEIOU'
    vowel_count = 0
    consonent_count = 0
    for a in s:
        if a in vowels:
            vowel_count += 1
        elif a.isalpha():
            consonent_count += 1
    return vowel_count, consonent_count

# Time: O(n) - single pass through string
# Space: O(1) - only counters

print(countVowels('Hello worlds !'))
print(countVowelsConsonants("Hello World!"))
```

### Optimized Solutions

```python
# Using sum() with generator
def count_vc(s):
    vowels = set('aeiouAEIOU')
    v = sum(1 for c in s if c in vowels)
    c = sum(1 for c in s if c.isalpha() and c not in vowels)
    return v, c

print(count_vc('Hello World!'))

# One-liner
count = lambda s: (sum(c.lower() in 'aeiou' for c in s if c.isalpha()), sum(c.isalpha() and c.lower() not in 'aeiou' for c in s))
print(count('Hello World!'))

# Time: O(n), Space: O(1)
```

### 🔑 Key Insights:

| Aspect | Your Code | Optimized |
|--------|-----------|-----------|
| Vowel lookup | String `in` | `set` (faster) |
| Approach | Two pointers | Single pass |
| Pythonic | Good | Generator expressions |

**Performance tip**: `set` lookup is O(1), string `in` is O(k) where k = vowel string length

**Your best**: `countVowels()` - clean and efficient! 🎯

13. **Reversing a String**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, String Manipulation  
    **Description**: Write a program to reverse a given string.  
    **Example**:  
    Input: `string = "programming"`  
    Output: `"gnimmargorp"`  
    Explanation: The reversed string of "programming" is "gnimmargorp". 

### 📝 Your Solutions vs Optimized Comparison

```python
# ========== YOUR SOLUTIONS (Strings) ==========

# Your Solution 1: Reverse iteration
def reverseString(s):
    rev=""
    for i in range(len(s)-1,-1,-1):
        rev += s[i]  # ⚠️ String concatenation creates new string each time
    return rev

# Time: O(n²) - string concatenation is O(n) per iteration!
# Space: O(n) - new string of length n
    
print(reverseString("hello"))

# Your Solution 2: Slicing ⭐ BEST!
def reverseString2(s):
    return s[::-1]

# Time: O(n) - single pass
# Space: O(n) - new reversed string
    
print(reverseString2("what"))

# Your Solution 3: Prepend each character
def reverseString3(s):
    rev=""
    for i in s:
        rev = i + rev  # ⚠️ Also O(n) per iteration
    return rev

# Time: O(n²) - string concatenation overhead
# Space: O(n)

print(reverseString3("python"))
```

### Optimized Solutions

```python
# Best: Slicing
reverse = lambda x: x[::-1]
print(reverse('hello'))
print(reverse([1,2,3]))

# Using reversed()
print(''.join(reversed('hello')))
print(list(reversed([1,2,3])))

# In-place for lists
lst = [1,2,3]
lst.reverse()
print(lst)

# Time: O(n), Space: O(n) for new, O(1) for in-place
```

```python
# ========== YOUR SOLUTIONS (Lists) ==========

# Your Solution 1: List concatenation
def reverselist(s):
    rev=[]
    for i in range(len(s)):
        rev = [s[i]] + rev  # ⚠️ Creates new list each time
    return rev

# Time: O(n²) - list concatenation is O(n) per iteration
# Space: O(n)
    
print(reverselist([1,2,3,4]))

# Your Solution 2: Slicing ⭐ BEST!
def reverselist2(s):
    return s[::-1]

# Time: O(n) - single pass
# Space: O(n) - new reversed list
    
print(reverselist2([1,2,3,4]))

# Your Solution 3: Prepend with concatenation
def reverselist3(s):
    rev= []
    for i in s:
        rev = [i]+rev  # ⚠️ O(n) per iteration
    return rev

# Time: O(n²)
# Space: O(n)

print(reverselist3([4,5,6,7,8]))

# Your Solution 4: Insert at beginning
def reverselist4(s):
    rev = []
    for i in s:
        rev.insert(0,i)  # ⚠️ insert(0) shifts all elements - O(n)
    return rev

# Time: O(n²) - insert(0) is O(n) per iteration
# Space: O(n)

print(reverselist4([1,2,3,4,5]))
```

### Key Insights

**Why O(n²)?**
- String `+=` creates new string each time (immutable)
- List `[x] + list` creates new list
- `insert(0)` shifts all elements

**Best:** Use `[::-1]` or `reversed()` or `.reverse()` for in-place

**Your best solution**: `reverseString2()` and `reverselist2()` - Perfect! 🎯
