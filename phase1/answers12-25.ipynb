13. **Reversing a String**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, String Manipulation  
    **Description**: Write a program to reverse a given string.  
    **Example**:  
    Input: `string = "programming"`  
    Output: `"gnimmargorp"`  
    Explanation: The reversed string of "programming" is "gnimmargorp". 

### 📝 Your Solutions vs Optimized Comparison

```python
# ========== YOUR SOLUTIONS (Strings) ==========

# Your Solution 1: Reverse iteration
def reverseString(s):
    rev=""
    for i in range(len(s)-1,-1,-1):
        rev += s[i]  # ⚠️ String concatenation creates new string each time
    return rev

# Time: O(n²) - string concatenation is O(n) per iteration!
# Space: O(n) - new string of length n
    
print(reverseString("hello"))

# Your Solution 2: Slicing ⭐ BEST!
def reverseString2(s):
    return s[::-1]

# Time: O(n) - single pass
# Space: O(n) - new reversed string
    
print(reverseString2("what"))

# Your Solution 3: Prepend each character
def reverseString3(s):
    rev=""
    for i in s:
        rev = i + rev  # ⚠️ Also O(n) per iteration
    return rev

# Time: O(n²) - string concatenation overhead
# Space: O(n)

print(reverseString3("python"))
```

### Optimized Solutions

```python
# Best: Slicing
reverse = lambda x: x[::-1]
print(reverse('hello'))
print(reverse([1,2,3]))

# Using reversed()
print(''.join(reversed('hello')))
print(list(reversed([1,2,3])))

# In-place for lists
lst = [1,2,3]
lst.reverse()
print(lst)

# Time: O(n), Space: O(n) for new, O(1) for in-place
```

```python
# ========== YOUR SOLUTIONS (Lists) ==========

# Your Solution 1: List concatenation
def reverselist(s):
    rev=[]
    for i in range(len(s)):
        rev = [s[i]] + rev  # ⚠️ Creates new list each time
    return rev

# Time: O(n²) - list concatenation is O(n) per iteration
# Space: O(n)
    
print(reverselist([1,2,3,4]))

# Your Solution 2: Slicing ⭐ BEST!
def reverselist2(s):
    return s[::-1]

# Time: O(n) - single pass
# Space: O(n) - new reversed list
    
print(reverselist2([1,2,3,4]))

# Your Solution 3: Prepend with concatenation
def reverselist3(s):
    rev= []
    for i in s:
        rev = [i]+rev  # ⚠️ O(n) per iteration
    return rev

# Time: O(n²)
# Space: O(n)

print(reverselist3([4,5,6,7,8]))

# Your Solution 4: Insert at beginning
def reverselist4(s):
    rev = []
    for i in s:
        rev.insert(0,i)  # ⚠️ insert(0) shifts all elements - O(n)
    return rev

# Time: O(n²) - insert(0) is O(n) per iteration
# Space: O(n)

print(reverselist4([1,2,3,4,5]))
```

### Key Insights

**Why O(n²)?**
- String `+=` creates new string each time (immutable)
- List `[x] + list` creates new list
- `insert(0)` shifts all elements

**Best:** Use `[::-1]` or `reversed()` or `.reverse()` for in-place

**Your best solution**: `reverseString2()` and `reverselist2()` - Perfect! 🎯

17. **Checking for Armstrong Numbers in a Range**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, Number Theory  
    **Description**: Write a program to find all Armstrong numbers within a given range.  
    **Example**:  
    Input: `range = [1, 500]`  
    Output: `[1, 153, 370, 371, 407]`  
    Explanation: Armstrong numbers between 1 and 500 are 1, 153, 370, 371, and 407. 

### 📝 Your Solutions vs Optimized Comparison

```python
# ========== YOUR SOLUTIONS ==========

# Your Solution 1: String conversion ⭐ BEST!
def armstrong(num):
    digit = str(num)
    res = 0
    for ch in digit:
        res  += int(ch) ** len(digit)
    return num == res

# Time: O(d) - where d = number of digits
# Space: O(d) - string storage

print(armstrong(153))

# Your Solution 2: Mathematical approach
def armstrong2(num):
    check = num
    length = len(str(num))  # ⚠️ Still uses string for length
    res = 0
    while num != 0:
        digit = num%10
        res += digit**length
        num = num//10
    return check == res

# Time: O(d) - d iterations for d digits
# Space: O(1) - only variables

print(armstrong2(153))  

# Your Solution 3: Range finder
def armstrong_range(start, end):
    res = []
    for num in range(start,end+1):
        digit = str(num)
        total = 0
        for ch in digit:
            total += int(ch) ** len(digit)
        if total == num:
            res.append(num)
    return res

# Time: O(n × d) - n numbers, d digits each
# Space: O(k) - k Armstrong numbers found

print(armstrong_range(1, 500))
```

### Optimized Solutions

```python
# One-liner check
is_armstrong = lambda n: n == sum(int(d)**len(str(n)) for d in str(n))
print(is_armstrong(153))  # True
print(is_armstrong(123))  # False

# List comprehension for range
def find_armstrong(start, end):
    return [n for n in range(start, end+1) 
            if n == sum(int(d)**len(str(n)) for d in str(n))]

print(find_armstrong(1, 500))

# Using filter
def armstrong_filter(start, end):
    return list(filter(lambda n: n == sum(int(d)**len(str(n)) for d in str(n)), 
                      range(start, end+1)))

print(armstrong_filter(1, 500))

# Optimized: Avoid repeated len() calls
def is_armstrong_optimized(n):
    digits = str(n)
    power = len(digits)
    return n == sum(int(d)**power for d in digits)

print(is_armstrong_optimized(9474))  # True

# Time: O(d) for check, O(n×d) for range
# Space: O(1) for check, O(k) for range
```
