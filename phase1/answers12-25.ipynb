17. **Checking for Armstrong Numbers in a Range**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, Number Theory  
    **Description**: Write a program to find all Armstrong numbers within a given range.  
    **Example**:  
    Input: `range = [1, 500]`  
    Output: `[1, 153, 370, 371, 407]`  
    Explanation: Armstrong numbers between 1 and 500 are 1, 153, 370, 371, and 407. 

### 📝 Your Solutions vs Optimized Comparison

````python
# ========== YOUR SOLUTIONS ==========

# Your Solution 1: String conversion ⭐ BEST!
def armstrong(num):
    digit = str(num)
    res = 0
    for ch in digit:
        res  += int(ch) ** len(digit)
    return num == res

# Time: O(d) - where d = number of digits
# Space: O(d) - string storage

print(armstrong(153))

# Your Solution 2: Mathematical approach
def armstrong2(num):
    check = num
    length = len(str(num))  # ⚠️ Still uses string for length
    res = 0
    while num != 0:
        digit = num%10
        res += digit**length
        num = num//10
    return check == res

# Time: O(d) - d iterations for d digits
# Space: O(1) - only variables

print(armstrong2(153))  

# Your Solution 3: Range finder
def armstrong_range(start, end):
    res = []
    for num in range(start,end+1):
        digit = str(num)
        total = 0
        for ch in digit:
            total += int(ch) ** len(digit)
        if total == num:
            res.append(num)
    return res

# Time: O(n × d) - n numbers, d digits each
# Space: O(k) - k Armstrong numbers found

print(armstrong_range(1, 500))
````

### Optimized Solutions

````python
# One-liner check
is_armstrong = lambda n: n == sum(int(d)**len(str(n)) for d in str(n))
print(is_armstrong(153))  # True
print(is_armstrong(123))  # False

# List comprehension for range
def find_armstrong(start, end):
    return [n for n in range(start, end+1) 
            if n == sum(int(d)**len(str(n)) for d in str(n))]

print(find_armstrong(1, 500))

# Using filter
def armstrong_filter(start, end):
    return list(filter(lambda n: n == sum(int(d)**len(str(n)) for d in str(n)), 
                      range(start, end+1)))

print(armstrong_filter(1, 500))

# Optimized: Avoid repeated len() calls
def is_armstrong_optimized(n):
    digits = str(n)
    power = len(digits)
    return n == sum(int(d)**power for d in digits)

print(is_armstrong_optimized(9474))  # True

# Time: O(d) for check, O(n×d) for range
# Space: O(1) for check, O(k) for range
````

    ```
    4 x 1 = 4  
    4 x 2 = 8  
    4 x 3 = 12  
    4 x 4 = 16  
    4 x 5 = 20  

````python
# Original multiplication function
def multiplication(num):
    for i in range(1,11):
        print(f"{num} x {i} = {num*i}")

multiplication(5)

# Complex version with additional features
def complex_multiplication(num, start=1, end=10, format_type="standard"):
    """
    Enhanced multiplication table generator with multiple formatting options
    
    Args:
        num: Number to generate multiplication table for
        start: Starting multiplier (default: 1)
        end: Ending multiplier (default: 10)
        format_type: Output format - "standard", "fancy", "aligned"
    
    Returns:
        List of multiplication results
    """
    results = []
    
    if format_type == "fancy":
        print(f"{'='*30}")
        print(f"Multiplication Table for {num}")
        print(f"{'='*30}")
    
    for i in range(start, end + 1):
        result = num * i
        results.append(result)
        
        if format_type == "standard":
            print(f"{num} x {i} = {result}")
        elif format_type == "fancy":
            print(f"│ {num:2d} × {i:2d} = {result:3d} │")
        elif format_type == "aligned":
            print(f"{num:3d} × {i:2d} = {result:4d}")
    
    if format_type == "fancy":
        print(f"{'='*30}")
    
    return results

# Usage examples of complex version
print("Complex version - Standard format:")
complex_multiplication(5)

print("\nComplex version - Fancy format:")
complex_multiplication(7, format_type="fancy")

print("\nComplex version - Custom range:")
complex_multiplication(3, start=5, end=15, format_type="aligned")

# Optimized solution using list comprehension and generator
def optimized_multiplication(num, start=1, end=10):
    """
    Memory-efficient multiplication table using generator
    """
    return (num * i for i in range(start, end + 1))

def print_optimized_table(num, start=1, end=10):
    """
    Print multiplication table using optimized generator approach
    """
    results = optimized_multiplication(num, start, end)
    for i, result in enumerate(results, start):
        print(f"{num} x {i} = {result}")

print("\nOptimized version:")
print_optimized_table(5)

# One-liner pythonic approach
def oneliner_multiplication(num: int, end: int = 10) -> None:
    """Ultra-concise multiplication table"""
    [print(f"{num} × {i} = {num * i}") for i in range(1, end + 1)]

print("\nOne-liner version:")
oneliner_multiplication(5)
````

````python
# Complex version with statistics
def complex_prime_finder(start, end, return_count=False, show_stats=False):
    primes = []
    for num in range(start, end + 1):
        if num <= 1:
            continue
        if num <= 3:
            primes.append(num)
            continue
        if num % 2 == 0 or num % 3 == 0:
            continue
        i = 5
        is_prime = True
        while i * i <= num:
            if num % i == 0 or num % (i + 2) == 0:
                is_prime = False
                break
            i += 6
        if is_prime:
            primes.append(num)
    
    if show_stats:
        print(f"Found {len(primes)} primes between {start} and {end}")
    print(primes)
    return (primes, len(primes)) if return_count else primes

print("Complex version:")
complex_prime_finder(10, 30, show_stats=True)

# Optimized - Sieve of Eratosthenes
def sieve_of_eratosthenes(start, end):
    if end < 2:
        return []
    sieve = [True] * (end + 1)
    sieve[0] = sieve[1] = False
    for i in range(2, int(end**0.5) + 1):
        if sieve[i]:
            for j in range(i*i, end + 1, i):
                sieve[j] = False
    return [i for i in range(start, end + 1) if sieve[i]]

print("\nOptimized (Sieve):")
print(sieve_of_eratosthenes(10, 30))

# Pythonic solution
def pythonic_primes(start: int, end: int) -> list[int]:
    def is_prime(n):
        if n < 2: return False
        if n == 2: return True
        if n % 2 == 0: return False
        return all(n % i != 0 for i in range(3, int(n**0.5) + 1, 2))
    return [n for n in range(start, end + 1) if is_prime(n)]

print("\nPythonic version:")
print(pythonic_primes(10, 30))
````

````python
# Complex version with validation
def complex_perfect_number(num, show_divisors=False):
    if num <= 0:
        return False
    divisors = [i for i in range(1, num) if num % i == 0]
    divisor_sum = sum(divisors)
    is_perfect = divisor_sum == num
    
    if show_divisors:
        print(f"Divisors of {num}: {divisors}")
        print(f"Sum: {divisor_sum}")
    print(f"{num} is {'a Perfect Number' if is_perfect else 'NOT a Perfect Number'}")
    return is_perfect

print("Complex version:")
complex_perfect_number(28, show_divisors=True)
complex_perfect_number(12, show_divisors=True)

# Optimized - sqrt optimization
def optimized_perfect_number(num):
    if num <= 1:
        return False
    divisor_sum = 1
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            divisor_sum += i
            if i != num // i:
                divisor_sum += num // i
    return divisor_sum == num

print("\nOptimized version:")
print(f"28: {optimized_perfect_number(28)}")
print(f"496: {optimized_perfect_number(496)}")

# Pythonic solution
pythonic_perfect = lambda n: n > 1 and sum(i for i in range(1, n) if n % i == 0) == n

print("\nPythonic version:")
print(f"28: {pythonic_perfect(28)}")
print(f"6: {pythonic_perfect(6)}")
````

23. **Finding the Fibonacci Number at a Specific Position**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, Sequences  
    **Description**: Write a program to find the Fibonacci number at a specific position.  
    **Example**:  
    Input: `position = 5`  
    Output: `5`  
    Explanation: The Fibonacci number at position 5 is 5 (sequence: 0, 1, 1, 2, 3, 5). 

````python
# Your Solution 1: Recursive (simple but inefficient)
def fibonacci(n):
    if n<=1:
        return n
    return fibonacci(n-1)+fibonacci(n-2)

# Time: O(2^n) - exponential, recalculates same values
# Space: O(n) - recursion stack depth

for i in range(6):
    print(fibonacci(i))
    
# Your Solution 2: Memoization ⭐ Good optimization!
def fibonacci_memo(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fibonacci_memo(n-1, memo) + fibonacci_memo(n-2, memo)
    return memo[n]

# Time: O(n) - each value calculated once
# Space: O(n) - memo dict + recursion stack

# Your Solution 3: Iterative ⭐ BEST!
def fibonacci_iter(n):
    a, b = 0,1
    for _ in range(n):
        a, b = b,a+b
    return a

# Time: O(n) - single loop
# Space: O(1) - only two variables

print(fibonacci_iter(10))

# Your Solution 4: Generate sequence
def fibonacci2(n):
    l = []
    a, b = 0,1
    for _ in range(n+1):
        l.append(a)
        a,b = b, b+a
    return l

# Time: O(n)
# Space: O(n) - stores all values

print(fibonacci2(5))
````

### Optimized & Pythonic Solutions

````python
# Optimized: Using lru_cache decorator
from functools import lru_cache

@lru_cache(maxsize=None)
def fib_cached(n):
    if n <= 1: return n
    return fib_cached(n-1) + fib_cached(n-2)

# Time: O(n), Space: O(n)
print(fib_cached(10))  # 55

# Pythonic: Generator for sequence
def fib_gen(n):
    a, b = 0, 1
    for _ in range(n + 1):
        yield a
        a, b = b, a + b

# Time: O(n), Space: O(1)
print(list(fib_gen(5)))  # [0, 1, 1, 2, 3, 5]

# One-liner using reduce
from functools import reduce
fib = lambda n: reduce(lambda x, _: (x[1], x[0] + x[1]), range(n), (0, 1))[0]
print(fib(10))  # 55

# Matrix exponentiation (O(log n))
def fib_matrix(n):
    def matrix_mult(a, b):
        return [[a[0][0]*b[0][0] + a[0][1]*b[1][0], a[0][0]*b[0][1] + a[0][1]*b[1][1]],
                [a[1][0]*b[0][0] + a[1][1]*b[1][0], a[1][0]*b[0][1] + a[1][1]*b[1][1]]]
    def matrix_pow(m, n):
        if n == 1: return m
        if n % 2 == 0:
            half = matrix_pow(m, n // 2)
            return matrix_mult(half, half)
        return matrix_mult(m, matrix_pow(m, n - 1))
    if n <= 1: return n
    return matrix_pow([[1, 1], [1, 0]], n)[0][1]

# Time: O(log n), Space: O(log n)
print(fib_matrix(10))  # 55
````

24. **Printing Prime Numbers Less Than a Given Number**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, Number Theory  
    **Description**: Write a program to print all prime numbers less than a given number.  
    **Example**:  
    Input: `number = 20`  
    Output: `2, 3, 5, 7, 11, 13, 17, 19`  
    Explanation: The prime numbers less than 20 are 2, 3, 5, 7, 11, 13, 17, and 19.

````python
# Your Solution 1: Find nth prime (inefficient)
def primeNumber(n):
    count = 0
    num = 1
    while count < n:
        num += 1
        for i in range(2,num):
            if num%i == 0:
                break
        else:
            count += 1
    return num

# Time: O(n × m²) - n primes, m is value, checking all divisors
# Space: O(1)

print(primeNumber(10))

# Your Solution 2: Optimized prime check ⭐ BEST!
def primenum(n):
    if n <= 1:
        return False
    if n<=3:
        return True
    if n%2== 0 or n%3 == 0:
        return False
    i = 5
    while i*i <=n :
        if n%i == 0 or n%i+2 ==0:
            return False
        i+=6
    return True

# Time: O(√n) - checks up to square root
# Space: O(1)

for i in range(20):
    if primenum(i):
        print(i)

````

### Optimized & Pythonic Solutions

````python
# Optimized: Sieve of Eratosthenes
def sieve(n):
    if n < 2: return []
    primes = [True] * n
    primes[0] = primes[1] = False
    for i in range(2, int(n**0.5) + 1):
        if primes[i]:
            for j in range(i*i, n, i):
                primes[j] = False
    return [i for i in range(n) if primes[i]]

# Time: O(n log log n) - most efficient for range
# Space: O(n)
print(sieve(20))  # [2, 3, 5, 7, 11, 13, 17, 19]

# Pythonic: List comprehension with helper
def is_prime(n):
    if n < 2: return False
    if n == 2: return True
    if n % 2 == 0: return False
    return all(n % i for i in range(3, int(n**0.5) + 1, 2))

primes = [i for i in range(20) if is_prime(i)]
print(primes)  # [2, 3, 5, 7, 11, 13, 17, 19]

# One-liner using filter
print(list(filter(lambda n: n > 1 and all(n % i for i in range(2, int(n**0.5) + 1)), range(20))))

# Generator for memory efficiency
def prime_gen(limit):
    yield 2
    for n in range(3, limit, 2):
        if all(n % i for i in range(3, int(n**0.5) + 1, 2)):
            yield n

print(list(prime_gen(20)))  # [2, 3, 5, 7, 11, 13, 17, 19]
````

25. **Finding the Number of Digits in a Number**  
    **Difficulty**: Easy  
    **Topics**: Basic Programming, Mathematical Computations  
    **Description**: Write a program to count the number of digits in a given number.  
    **Example**:  
    Input: `number = 12345`  
    Output: `5`  
    Explanation: The number 12345 has 5 digits.  

````python
# Your Solution 1: String conversion ⭐ BEST!
def count(num):
    return len(str(num))

# Time: O(d) - d is number of digits
# Space: O(d) - string storage

print(count(1234))

# Your Solution 2: Mathematical approach
def count2(num):
    count = 0
    while num > 0:
        count += 1
        num = num//10
    return count

# Time: O(d) - d iterations
# Space: O(1) - only counter

print(count2(12345))
````

### Optimized & Pythonic Solutions

````python
# Optimized: Using logarithm O(1)
import math

def count_log(num):
    if num == 0: return 1
    return int(math.log10(abs(num))) + 1

# Time: O(1) - constant time
# Space: O(1)
print(count_log(12345))  # 5
print(count_log(0))  # 1

# Pythonic: One-liner
count_digits = lambda n: len(str(abs(n)))
print(count_digits(12345))  # 5
print(count_digits(-999))  # 3

# Handle negative numbers
def count_all(num):
    return len(str(abs(num)))

print(count_all(-12345))  # 5

# Using format
print(len(f'{12345}'))  # 5

# Recursive approach
def count_recursive(n):
    if n == 0: return 0
    return 1 + count_recursive(n // 10)

# Time: O(d), Space: O(d) - recursion stack
print(count_recursive(12345))  # 5
````
